```python
import pandas as pd
from os import listdir
import pickle
import matplotlib.pyplot as plt

def estimation_summary(res, target_recall=.9, target_cost=.2, filename=''):
    df = pd.DataFrame(res['supervised'])  
    ground_truth = res['true']
    ground_truth

    try:
        clf_df = df.loc[df.x >=df.thres].iloc[0]
        sup_pos = clf_df['pos']
        sup_est = clf_df['est']
        sup_cost = clf_df['x']
    except:
        print("No Thres")
        
    recall_90_df = df.loc[df.est*.9 <= df.pos].iloc[0]
    stop_90_pos = recall_90_df['pos']
    stop_90_est = recall_90_df['est']
    stop_90_cost = recall_90_df['x']
    
    recall_95_df = df.loc[df.est*.95 <= df.pos].iloc[0]
    stop_95_pos = recall_95_df['pos']
    stop_95_est = recall_95_df['est']
    stop_95_cost = recall_95_df['x']


    five_df = df.loc[df.x >= (ground_truth[1] * .05)].iloc[0]
    five_pos = five_df['pos']
    five_est = five_df['est']
    five_cost = five_df['x']
    
    ten_df = df.loc[df.x >= (ground_truth[1] * .1)].iloc[0]
    ten_pos = ten_df['pos']
    ten_est = ten_df['est']
    ten_cost = ten_df['x']
    
    twenty_df = df.loc[df.x >= (ground_truth[1] * .2)].iloc[0]
    twenty_pos = twenty_df['pos']
    twenty_est = twenty_df['est']
    twenty_cost = twenty_df['x']
    
    thirty_df = df.loc[df.x >= (ground_truth[1] * .3)].iloc[0]
    thirty_pos = thirty_df['pos']
    thirty_est = thirty_df['est']
    thirty_cost = thirty_df['x']
    
    forty_df = df.loc[df.x >= (ground_truth[1] * .4)].iloc[0]
    forty_pos = forty_df['pos']
    forty_est = forty_df['est']
    forty_cost = forty_df['x']
    
    fifty_df = df.loc[df.x >= (ground_truth[1] * .5)].iloc[0]
    fifty_pos = fifty_df['pos']
    fifty_est = fifty_df['est']
    fifty_cost = fifty_df['x']
    
    
    print("FILENAME: " + filename)
    print("TRUE:", ground_truth)
    try:
        print("CLF:")
        print("True recall: %.4f, est recall: %.4f, precision: %.4f" %(sup_pos/ground_truth[0], sup_pos/sup_est, sup_pos/float(sup_cost)))
    except:
        print("No Thres")
    
    print("Target Cost %.2f:" %.05)
    print("Est: %.2f, Accu: %.2f, Read: %.2f, Cost: %.2f, Found: %.2f, Recall: %.2f" %(five_est, five_est/ground_truth[0]*100, 
                                                                                      five_cost, five_cost/ground_truth[1]*100, 
                                                                                      five_pos, five_pos/ground_truth[0]*100))
    
    print("Target Cost %.2f:" %.1)
    print("Est: %.2f, Accu: %.2f, Read: %.2f, Cost: %.2f, Found: %.2f, Recall: %.2f" %(ten_est, ten_est/ground_truth[0]*100, 
                                                                                      ten_cost, ten_cost/ground_truth[1]*100, 
                                                                                      ten_pos, ten_pos/ground_truth[0]*100))
    
    
    print("Target Cost %.2f:" %.2)
    print("Est: %.2f, Accu: %.2f, Read: %.2f, Cost: %.2f, Found: %.2f, Recall: %.2f" %(twenty_est, twenty_est/ground_truth[0]*100, 
                                                                                      twenty_cost, twenty_cost/ground_truth[1]*100, 
                                                                                      twenty_pos, twenty_pos/ground_truth[0]*100))
    
    print("Target Cost %.2f:" %.3)
    print("Est: %.2f, Accu: %.2f, Read: %.2f, Cost: %.2f, Found: %.2f, Recall: %.2f" %(thirty_est, thirty_est/ground_truth[0]*100, 
                                                                                      thirty_cost, thirty_cost/ground_truth[1]*100, 
                                                                                      thirty_pos, thirty_pos/ground_truth[0]*100))
    
    print("Target Cost %.2f:" %.4)
    print("Est: %.2f, Accu: %.2f, Read: %.2f, Cost: %.2f, Found: %.2f, Recall: %.2f" %(forty_est, forty_est/ground_truth[0]*100, 
                                                                                      forty_cost, forty_cost/ground_truth[1]*100, 
                                                                                      forty_pos, forty_pos/ground_truth[0]*100))
    
    print("Target Cost %.2f:" %.5)
    print("Est: %.2f, Accu: %.2f, Read: %.2f, Cost: %.2f, Found: %.2f, Recall: %.2f" %(fifty_est, fifty_est/ground_truth[0]*100, 
                                                                                      fifty_cost, fifty_cost/ground_truth[1]*100, 
                                                                                      fifty_pos, fifty_pos/ground_truth[0]*100))
    
    
    print("Target Recall %.2f:" %.9)
    print("Est: %.2f, Accu: %.2f, Read: %.2f, Cost: %.2f, Found: %.2f, Recall: %.2f" %(stop_90_est, stop_90_est/ground_truth[0]*100, 
                                                                                      stop_90_cost, stop_90_cost/ground_truth[1]*100, 
                                                                                      stop_90_pos, stop_90_pos/ground_truth[0]*100))    
    
    print("Target Recall %.2f:" %.95)
    print("Est: %.2f, Accu: %.2f, Read: %.2f, Cost: %.2f, Found: %.2f, Recall: %.2f" %(stop_95_est, stop_95_est/ground_truth[0]*100, 
                                                                                      stop_95_cost, stop_95_cost/ground_truth[1]*100, 
                                                                                      stop_95_pos, stop_95_pos/ground_truth[0]*100))    
    
    print("TARGET90 GAIN:")
    print("max: %.2f, current: %.2f" %(1/(ground_truth[0]*.9/ground_truth[1]), (stop_90_pos/ground_truth[0])/(stop_90_cost/ground_truth[1])))
    print(stop_90_pos/ground_truth[0])
    print(stop_90_cost/ground_truth[1])
    print("For LATEX")
    print('', int(round(five_pos/ground_truth[0]*100,0)), ' &\n', int(round(ten_pos/ground_truth[0]*100,0)), ' &\n', 
          int(round(twenty_pos/ground_truth[0]*100,0)), ' &\n', int(round(thirty_pos/ground_truth[0]*100,0)), ' &\n',
          int(round(forty_pos/ground_truth[0]*100,0)), ' &\n', int(round(fifty_pos/ground_truth[0]*100,0)), '&\n', 
          int(round(stop_90_cost/ground_truth[1]*100,0)), '&', int(round(stop_90_pos/ground_truth[0]*100,0)), '&\n', 
          int(round(stop_95_cost/ground_truth[1]*100,0)), '&', int(round(stop_95_pos/ground_truth[0]*100,0)), '\\\\')
    
    results = []
    results_cost = []
    results.append(round(five_pos/ground_truth[0]*100,0))
    results.append(round(ten_pos/ground_truth[0]*100,0))
    results.append(round(twenty_pos/ground_truth[0]*100,0))
    results.append(round(thirty_pos/ground_truth[0]*100,0))
    results.append(round(forty_pos/ground_truth[0]*100,0))
    results.append(round(fifty_pos/ground_truth[0]*100,0))
    results.append(round(stop_90_pos/ground_truth[0]*100,0)) #recall
    results.append(round(stop_90_cost/ground_truth[1]*100,0))
    results.append(round(stop_95_pos/ground_truth[0]*100,0)) #recall
    results.append(round(stop_95_cost/ground_truth[1]*100,0))

    print('\n')
    return results_cost, results
    
files = listdir('../../dump_90_boost_dt_combined')
df = pd.DataFrame()
for file in files:
    with open('../../dump_90_boost_dt_combined/' + file, "rb") as handle:
        res = pickle.load(handle)
        filename = file.rsplit('.',1)[0]
        print(filename) 
        results_cost, results_target = estimation_summary(res, filename=filename)
        df[filename] = results_target
df = df.T
cols = ['5', '10', '20', '30', '40', '50', 'T90Recall', 'T90Cost', 'T95Recall', 'T95Cost']
df.columns = cols
#df = df.T


print(df.describe().loc['75%'] - df.describe().loc['25%'])
df.describe()
```
